# 판다스로 위키페이지 웹크롤링 하기

```python
# 라이브러리 로드

import pandas as pd
```

```python
# 크로링할 웹페이지를 url 변수로 저장 
url = "https://ko.wikipedia.org/wiki/%EB%8C%80%ED%95%9C%EB%AF%BC%EA%B5%AD%EC%9D%98_%EC%9D%B8%EA%B5%AC"
url
```

```python
# pandas로 위키피디아 웹페이지 읽어오기 

table = pd.read_html(url) # html문서의 테이블을 가져옴 


len(table) # len으로 4개의 테이블이 존재한다는 것을 알 수 있음
```

```python
table[0]
```

```python
# 분석하고자하는 2번째 테이블을 df로 저장 
df = table[1] 
df.shape
```

```python
df.head()
```

```python
df.tail()
```

# 대한민국 인구통계 pointplot, lineplot으로 그리기 

```python
# 라이브러리 로드
import matplotlib.pyplot as plt
import seaborn as sns 
sns.set(font="Malgun Gothic")
```

```python
# 폰트 선명하게 설정
from IPython.display import set_matplotlib_formats
set_matplotlib_formats("retina")
```

```python
# 추계인구(명) lineplot
plt.figure(figsize=(18,8)) # 그래프 사이즈 설정
plt.xticks(rotation = 60)  # x 축 글자 각도 설정 
sns.pointplot(data=df, x="연도 (년)", y="추계인구(명)")
```

```python
# lineplot으로 그리기
plt.figure(figsize=(18,8)) # 그래프 사이즈 설정
plt.xticks(rotation = 60)  # x 축 글자 각도 설정 
sns.lineplot(data=df, x="연도 (년)", y="추계인구(명)")
```

```python
# 출생자수(명) pointplot
plt.figure(figsize=(18,8)) # 그래프 사이즈 설정
plt.xticks(rotation = 60)  # x 축 글자 각도 설정 
sns.pointplot(data=df, x="연도 (년)", y="출생자수(명)")
```

```python
# 출생자수(명) lineplot
plt.figure(figsize=(18,8)) # 그래프 사이즈 설정

sns.lineplot(data=df, x="연도 (년)", y="출생자수(명)")
```

```python
# 사망자수(명) pointplot
plt.figure(figsize=(18,8)) # 그래프 사이즈 설정
plt.xticks(rotation = 60)  # x 축 글자 각도 설정 
sns.pointplot(data=df, x="연도 (년)", y="사망자수(명)")
```

# 대한민국 인구통계 출생자수, 사망자수 같은 그래프에 그리기

```python
# 출생자수, 사망자수 같이 그리기
plt.figure(figsize=(18,8)) # 그래프 사이즈 설정
plt.xticks(rotation = 60)  # x 축 글자 각도 설정 
sns.pointplot(data=df, x="연도 (년)", y="출생자수(명)")
sns.pointplot(data=df, x="연도 (년)", y="사망자수(명)", color="orange")
```

## 판다스에서 그래프 그리기

```python
# 필요한 컬럼만 가져오기 
df[['연도 (년)', '출생자수(명)', '사망자수(명)']] # 여러컬럼을 가져올때는 리스트 형태로 
```

```python
# 연도가 x축으로 오지 않고 y 값으로 들어가 있음
df[['연도 (년)', '출생자수(명)', '사망자수(명)']].plot()
```

```python
# 연도를 set_index를 사용해 인덱스로 할당
df_pop = df[['연도 (년)', '출생자수(명)', '사망자수(명)']]
df_pop = df_pop.set_index('연도 (년)')
df_pop
```

```python
# 그래프로 그리기 
df_pop.plot(figsize=(15,4))
```
