```python
#라이브러리 로드 
import pandas as pd
import numpy as np

# 그래프가 노트북 안에 보이게 하기 위해
%matplotlib inline
```

```python
#데이터 불러오기
df_kosis = pd.read_csv("data/출생아수_시도_시_군_구__20200703155657.csv", encoding="cp949")
df_kosis.shape
```

# 데이터 전처리

## tidy data 만들기

```python
# 분석 편의를 위한 tidy data 만들기 
df = df_kosis.melt(id_vars="시군구별")
df.head()
```

## 필요없는 행 제거 

```python
# 행들의 유니크 값 확인  
df["시군구별"].unique()
```

```python
# 값이 시군구별인 data 확인 
df[df["시군구별"] == "시군구별"]
```

```python
# 값이 시군구별인 data는 필요없는 값이기 때문에 제거 
df = df[df["시군구별"] != "시군구별"].copy()
df.shape
```

```python
df["시군구별"].unique()
```

```python
df.head()
```

## 연도, 월, 성별, 텍스트 데이터 전처리

```python
# variable 값 나눠주기 
df["variable"].str.split(".", expand=True) # expand=True 나눈 값을 데이터 프레임 형태로 반환 
```

```python
# 연도 컬럼 만들어주기 
df["연도"] = df["variable"].str.split(".", expand=True)[0]
df.head()
```

```python
# 월 컬럼 만들어주기 
df["월"] = df["variable"].str.split(".", expand=True)[1]
df.head()
```

```python
# 성별 컬럼 만들어주기 
df["성별"] = df["variable"].str.split(".", expand=True)[2]
df.head()
df.tail()
```

```python
df["성별"].unique()
```

```python
# unique 값의 갯수, None은 포함되지 않음 
df["성별"].nunique()
```

```python
# 성별의 None 값을 "전체"로 바꿔주기 
df["성별"] = df["성별"].fillna("전체")
df["성별"].unique()
```

```python
# 성별의 "1", "2" 값 "남", "여"로 바꿔주기 
df["성별"] = df["성별"].replace("1", "남").replace("2", "여")
df["성별"].unique()
```

```python
df["성별"].value_counts()
```

```python
df.head()
```

```python
# 컬럼이름 바꿔주기
df = df.rename(columns={"variable":"기간", "value":"출생아수"})
df.head()
```

#  출생아수 시각화

## 형변환

```python
df
```

```python
# "출생아수"가 object 타입으로 되어있어 형변환 
#df["출생아수"].astype(int)
```

```python
# "-" 때문에 int로 형변환이 안되기 때문에 nan으로 변환  
df["출생아수"] = df["출생아수"].replace("-", np.nan)
```

```python
# NaN은 float 타입이기 때문에 float으로 형변환 
df["출생아수"] = df["출생아수"].astype(float)
df["출생아수"].describe()
```

```python
# 시군구별 = 전국, 성별 = 전체 데이터 색인하여 가져오기 
df_all = df[(df["시군구별"] == "전국") & (df["성별"] == "전체")]
df_all
```

## 출생아수 pandas plot으로 그리기 

```python
# 필요한 컬럼만 추출
df_all = df_all[["연도", "월", "출생아수"]].copy()
df_all.head()
```

```python
# 폰트 설정
import matplotlib.pyplot as plt

plt.rc("font", family = "Malgun Gothic")
```

```python
# 연도와 월을 인덱스로 지정, 선그래프 그리기 
df_all.set_index(["연도", "월"]).plot(figsize=(15,4))
```

출생자 수에 계절성이 있음

```python
# 연도와 월을 인덱스로 지정,막대그래프, 슬라이싱으로 최근 48 개월로 
df_all[-48: ].set_index(["연도", "월"]).plot.bar(figsize=(15,4))
```

출생아 수가 1월, 3월이 많고 12월이 적다

## 시도별 출생아수 시각화

```python
df_local = df[df['시군구별']!='전국'].copy() # copy()는 깊은복사, 그냥 할당하면 옅은 복사, 옅은 복사에서는 df_local 값을 변경하면 
# df 값도 변화 한다. 

df_local.head()
```

```python
plt.figure(figsize=(15,5))
sns.pointplot(data=df_local, x="연도", y='출생아수', ci=None, hue='성별')
```

```python
df_local_all = df_local[df_local['성별']=='전체'].copy()
df_local_all
```

```python
plt.figure(figsize=(15,8))

sns.pointplot(data=df_local_all, x='연도', y='출생아수', ci=None, hue='시군구별')
plt.legend(bbox_to_anchor=(1.01, 1), loc=2, borderaxespad=0.) # legend 밖에 그리기 
```

출생아수가 전체적으로 줄고 있지만 세종시는 증가

```python
# 경기도, 서울, 세종특별자치시로 서브 데이터 형성
df_local_2 =  df_local_all[df_local_all['시군구별'].isin(['세종특별자치시', '서울특별시', '경기도'])].copy()
df_local_2
```
