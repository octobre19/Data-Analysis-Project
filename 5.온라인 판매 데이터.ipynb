# 분석과 시각화를 위한 tidy data 만들기


```python
df_raw.columns
```

```python
df = df_raw.melt(id_vars=["국가(대륙)별","상품군별", "판매유형별"], var_name="기간", value_name="백만원")

df.shape
```

```python
df.head()
```

```python
df.info()
```

# 데이터 전처리

## 기간에서 연도를 분리하기

```python
int("2019 4/4".split()[0])
```

```python
# 연도만 분리 
# map() 안에 함수 지정 가능
df["연도"] = df["기간"].map(lambda x : int(x.split()[0]))

df.head()
```

```python
# 기간에서 분기 만 분리 

"2019 1/4".split()[1].split("/")[0]
```

```python
# 기간에서 분기 만 분리 
df["분기"] = df["기간"].map(lambda x : int(x.split()[1].split("/")[0]))
df.head()
```

## 금액을 수치데이터로 표현하기 위해 데이터 타입 변경하기


```python
# - 문자를 결측치로 변경하고 float 타입으로 변경하기

df["백만원"] = df["백만원"].replace("-", pd.np.nan).astype(float) # 결측치 데이터가 flaot 타입이기 때문에 float으로 형 변환

df.head()
```

## 필요없는 데이터 제거하기 

```python
df.info()
```

```python
# 합계 데이터는 따로 구학 수 있기 때문에 전체 데이터에서 제거합니다. 

df = df[(df["국가(대륙)별"] != "합계") & (df["상품군별"] != "합계")].copy()
df.info()
```

```python
# 결측치 보기

df.isnull().sum()
```

# K-beauty 시각화

## 전체 상품군 판매액


```python
# 판매유형별 데이터는 일부 기간에는 "계" 만 존재하기 때문에 판매유형별 == "계" 데이터만 가져와서 봐야지
# 평균값을 구하는 그래프에서 올바른 값을 표현할 수 있습니다.

df_total = df[df["판매유형별"] == "계"].copy()

df_total.head()
```

```python
# 연도, 판매액 lineplot으로 그리기

sns.lineplot(data=df_total, x="연도", y="백만원")
```

```python
# 연도, 판매액 lineplot으로 그리고 상품군별로 다른 색상으로 표현하기



sns.lineplot(data=df_total,x="연도", y="백만원", hue="상품군별")
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
```

```python
# 위에 그른 그래프를 자세히 보기 위해 서브플롯으로 표시하기
sns.relplot(data=df_total,x="연도",y="백만원",
            hue="상품군별", kind="line", col="상품군별", col_wrap=4)
```

```python
# isin을 사용해 화장품만 제외하고 df_sub 이라는 변수에 담습니다.

df_sub = df_total[~df_total["상품군별"].isin(["화장품", "의류 및 패션관련 상품"])].copy()
```

```python
# 연도별 판매액을 상품군별로 relplot을 활용해 서브플롯으로 그려보기
sns.relplot(data = df_sub, x="연도", y="백만원", hue="상품군별",
           col="상품군별", col_wrap=4, kind="line")
```

## 화장품의 온라인쇼핑 해외직접판매액

```python
# 화장품 판매액에 대한 기간별 금액 데이터 시각화 gkrh "국가(대륙)별" 다른 색상으로 표시하기
# 중국만 빼고 표시
plt.figure(figsize=(15,4))
plt.xticks(rotation=30)
sns.lineplot(data=df_cosmetic[df_cosmetic["국가(대륙)별"] != "중국"], x="기간", y="백만원", hue="국가(대륙)별")
```

```python
# 화장품 판매액에 대한 기간별 금액 데이터를 시각화하고 "판매유형별"로 다른 색상으로 표시하기
plt.figure(figsize=(15,4))
plt.xticks(rotation=30)
df_sub = df[df["판매유형별"]!="계"].copy()
sns.lineplot(data=df_sub, x="기간", y="백만원", hue="판매유형별")
```

```python
# 화장품 판매액에 대한 기간별 금액 데이터를 시각화하고 "판매유형별"로 다른 색상으로 표시하기
plt.figure(figsize=(15,4))
plt.xticks(rotation=30)
df_sub = df[(df["판매유형별"]!="계") & (df["판매유형별"]!="면세점") ].copy()
sns.lineplot(data=df_sub, x="기간", y="백만원", hue="판매유형별", ci=None)
```

## 의류 및 패션관련 상품 온라인쇼핑 해외직접판매액


```python
# df_fashion이라는 변수에 의류 데이터만 가져와 따로 담아주기
df_fashion = df[(df["상품군별"] == "의류 및 패션관련 상품") & (df["판매유형별"] == "계")].copy()
df_fashion.head()
```

```python
# "의류 및 패션관련 상품" 판매액에 대한 기간별 금액 데이터 시각화
# "국가(대륙)별"로 다른 색상으로 표시

plt.figure(figsize=(15,4))
plt.xticks(rotation=30)
sns.lineplot(data=df_fashion, x="기간", y="백만원", hue="국가(대륙)별")
```

```python
# "의류 및 패션관련 상품" 판매액에 대한 기간별 금액 데이터를 시각화 하고
# "판매유형별"로 다른 색상으로 표시하기

df_fashion2 = df[(df["상품군별"] == "의류 및 패션관련 상품") & (df["판매유형별"] != "계")].copy()

plt.figure(figsize=(15,4))
plt.xticks(rotation=30)
sns.lineplot(data=df_fashion2, x="기간", y="백만원", hue="판매유형별", ci=None)
```
